```python
# Import PuLP
from pulp import *

# Create the problem instance
prob = LpProblem("Transportation Problem", LpMinimize)
```

### 2.1 Define data

```python
# Suppliers (Warehouses)
Warehouses = ["W1", "W2"]

# Demanders (Stores)
Stores = ["S1", "S2", "S3"]

# Supply of each warehouse
supply = {"W1": 100,
          "W2": 120}

# Demand of each store
demand = {"S1": 80,
          "S2": 70,
          "S3": 70}

# Unit transportation cost matrix (cij)
costs = {"W1": {"S1": 2, "S2": 4, "S3": 5},
         "W2": {"S1": 3, "S2": 1, "S3": 7}}
```

### 2.2 Define Decision Variables

```python
# Create the decision variables x_ij
# x_ij represents the quantity of goods transported from warehouse i to store j
x = LpVariable.dicts("x", [(i, j) for i in Warehouses for j in Stores], 0, None, LpInteger)

# Display the decision variables
print("Decision Variables:")
for var_name, var_obj in x.items():
    print(f"{var_name}: {var_obj}")
```

### 2.3 Define Objective Function

```python
# Add the objective function to the problem
prob += lpSum([x[(i, j)] * costs[i][j] for i in Warehouses for j in Stores]), "Total Cost"
```

### 2.4 Define Constraints

```python
# Supply constraints (warehouse capacity)
# Clear existing constraints to avoid 'overlapping constraint names' error on re-run
prob.constraints.clear()

for i in Warehouses:
    prob += lpSum([x[(i, j)] for j in Stores]) <= supply[i], f"Sum_of_products_out_of_warehouse_{i}"

# Demand constraints (store requirements)
for j in Stores:
    prob += lpSum([x[(i, j)] for i in Warehouses]) == demand[j], f"Sum_of_products_into_store_{j}"

# Print the problem to verify formulation
print("\nProblem Formulation:")
print(prob)
```

### 2.5 Solve the Problem

```python
# The problem is solved using PuLP's default solver
prob.solve()

# Print the status of the solution
print("\nStatus:", LpStatus[prob.status])
```

### 2.7 Visualization of the Optimal Solution

```python
# Print the optimal values of the decision variables
print("\nOptimal Solution:")
for v in prob.variables():
    if v.varValue > 0:
        print(f"{v.name} = {v.varValue}")

# Print the minimum total cost
print(f"\nMinimum Total Cost = {value(prob.objective)}")
```
